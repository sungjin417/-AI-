### 2-5. 데이터 필터링
1. 단일조건으로 데이터 선택
2. 여러 조건을 사용한 데이터 선택
3. 특정 값을 포함하는 데이터 선택
4. 특정 조건을 만족하는 행의 인덱스만 선택

```python
data = {
    '학생' : ['철수', '짱구', '훈이', '유리', '수지', '흰둥이', '짱아', '맹구', '미리', '성아'],
    '국어' : [88, 92, 76, 81, 95, 67, 80, 91, 85, 73],
    '영어' : [78, 85, 92, 88, 76, 84, 90, 81, 89, 80],
    '수학' : [90, 78, 82, 94, 87, 79, 83, 88, 76, 92],
    '사회' : [85, 92, 88, 77, 91, 82, 90, 79, 84, 88],
    '과학' : [91, 79, 88, 85, 82, 90, 87, 94, 83, 86]
}
df = pd.DataFrame(data)
df
```

```python
df['국어'] >= 80
```

```python
df[df['국어'] >= 80]
```

```python
# 국어점수가 80점 이상이고, 사회점수가 85점인 데이터
# and 조건 '&'
df[(df['국어'] >= 80) & (df['사회'] == 85)]
```

```python
# 영어점수가 85점 이상이거나, 수학 점수가 90점 이상인 데이터
# or 조건 '|', .isin()
df[(df['영어'] >= 85) | (df['수학'] >= 90)]
```

```python
df[(df['과학'] == 91) | (df['과학'] == 86)]
```

```python
# 과학점수가 91점, 86점인 데이터,
df[(df['과학']).isin([91,86])]
```

```python
# 영어점수가 90점 이상인 데이터의 인덱스만 알고 싶다 -> .index
df[df['영어'] >= 90].index
```

## 4. Groupby를 통한 데이터 필터링

### groupby
- groupby : 데이터프레임의 데이터를 그룹으로 묶고,각 그룹에 대해 특정 연산을 적용할 수 있도록 하는 기능
- 집계 함수(예: sum(), mean(), count(), max() 등)와 함께 자주 사용


```python
data = {
    '과일' : ['사과','사과', '바나나', '바나나', '포도', '포도'],
    '판매량' : [5, 3, 6, 7, 4, 2]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('과일')['판매량'].sum() # 과일별 판매량 총 합
```

```python
df.groupby('과일')['판매량'].mean() # 과일별 판매량 평균
```

```python
df.groupby('과일')['판매량'].max() # 과일별 판매량 최대값
```

```python
df.groupby('과일')['판매량'].count() # 과일별 판매량 row 개수
```

### 5-2. 데이터프레임 연결

```python
data1 = {
    '고객ID': [101, 102, 103, 104, 105],
    '이름': ['짱아', '철수', '짱구', '수지', '유리'],
    '구매금액': [25000, 30000, 25000, 50000, 25000],
    '구매횟수': [1, 2, 2, 4, 1]
}
df1 = pd.DataFrame(data1)
df1
```
