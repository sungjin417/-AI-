#### 1. RNN을 이용하여 텍스트 생성하기

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
text = """경마장에 있는 말이 뛰고 있다\n
그의 말이 법이다\n
가는 말이 고와야 오는 말이 곱다\n"""
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
sequences = []
for line in text.split('\n'):
    encoded = tokenizer.texts_to_sequences([line])[0]
    # print(encoded)
    for i in range(1, len(encoded)):
        # i = 1일 때, sequence = [2,3]
        # i = 2일 때, sequence = [2,3,1]
        sequence = encoded[:i+1]
        sequences.append(sequence)

print(f'학습에 사용할 샘플의 개수 : {len(sequences)}')
```

```python
sequences # 순차적으로 만들기
```

```python
# 문장의 길이를 동일하게 맞추기

lengths = []
for i in sequences:
    lengths.append(len(i))
```

```python
max_len = max(lengths)
print(f'샘플 중 최대 길이 : {max_len}')
```

```python
sequences = pad_sequences(sequences, maxlen=max_len, padding = 'pre') # pad_sequences : 길이를 맞추기 위해 패딩을 넣어주는 함수
```

```python
sequences
```

```python
sequences = np.array(sequences) # 오류 발생 할 수 있기 때문에 
```

```python
X = sequences[:, :-1]
y = sequences[:, -1]
```

```python
X
```

```python
y
```

```python
y = to_categorical(y, num_classes=vocab_size) # one-hot incoding
y
```

#### 2) 모델 설계하기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN # Embedding : 단어적 의미를 포함하고 있다

embedding_dim = 10
hidden_units = 32
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(vocab_size, activation='softmax'))
# word_index = 11, vector(padding) = 12, 
# vocab_size = 12인데 단어 11개 + 1 했던 word_index값
# +1인 자리차지를 위해 앞에 0을 채워준 것으로
# 실제 예측을 수행할 때에는 0은 단어가 아니므로 무시되어 출력 값으로 0이 나오지는 않는다

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, y, epochs=200, verbose=2)

```

```python
# 문장을 생성하는 함수 만들기
def sentence_generation(model, tokenizer, current_word, n):
    init_word = current_word # 사용자가 입력한 단어
    sentence = '' # 최종적으로 생성한 문장을 반환

    for i in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=5, padding='pre') # maxlen= : input과 output의 길이가 같아야 하기 때문에 5

        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis=1)

        for word, index in tokenizer.word_index.items():
            if index == result:
                break
            # (말이 ,1)
        current_word = current_word + ' ' + word
        # current_word는 다음 반복문에서 사용되기 위해 저장한 변수
        sentence = sentence + ' ' + word
        # sentence는 현재 출력을 위한 문장 생성
    sentence = init_word + sentence
    return sentence
```

```python
sentence_generation(model, tokenizer, '경마장에', 4)
```

```python
sentence_generation(model, tokenizer, '가는', 5)
```

### 2. LSTM을 이용하여 텍스트 생성하기

```python
from string import punctuation
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv('../data/ArticlesApril2018.csv')
df.head()
```

```python
print(f'열의 개수 : {len(df.columns)}')
print(df.columns)
```

```python
df['headline'].isnull().values.any()
```

```python
headline = []
headline.extend(list(df.headline.values))
headline[:5]
```

```python
print(f'총 샘플의 개수 (노이즈 제거 전) : {len(headline)}')
```

```python
clean_headline = []
for word in headline:
    if word != 'Unknown':
        clean_headline.append(word)
print(f'샘플의 개수 (노이즈 제거 후) : {len(clean_headline)}')
```

```python
clean_headline[:5]
```

#### 전처리

```python
# 전처리 함수
def repreprocessing(raw_sentence):
    preprocessed_sentence = raw_sentence.encode('utf8').decode('ascii', 'ignore')
    # 아스키 : 단순하고 영어 텍스트에 효과적
    # utf8 : 다국어와 기호 처리 가능
    # ignore : 아스키 문자가 아닌것들은 무시한다
    # 해당 데이터셋은 영어로 되어있기 때문에 우선적으로 다국어와 기호를 처리한 뒤
    # 단순하고 영어 텍스트에 효과적인 아스키 문자로 변환
    result = ''
    for word in preprocessed_sentence:
        if word not in punctuation: # 구두점 제거
            result += word
    return result.lower() # 소문자 화
```

```python
preprocessed_headline = []
# 전처리 된 헤드라인 담아주기
for word in clean_headline:
    preprocessed_sentence = repreprocessing(word)
    preprocessed_headline.append(preprocessed_sentence)

preprocessed_headline[:5]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_headline)
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(f'단어 집합의 크기 : {vocab_size}')
```

```python
sequences = []
for sentence in preprocessed_headline:
    encoded = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        # 인코딩된 문장이 [3, 5, 6, 7] 라면 
        # 첫번째 반복에서는 [3, 5] 
        # 두번째 반복에서는 [3, 5, 6]
        sequences.append(sequence)
```

```python
sequences[:11]
# 'former nfl cheerleaders settlement offer 1 and a meeting with goodell'
```

```python
lengths = []
for i in sequences:
    lengths.append(len(i))

max_len = max(lengths)
print(f'샘플의 최대 길이 {max_len}')
```

```python
sequences = pad_sequences(sequences, maxlen = max_len, padding = 'pre')
print(sequences[:3])
```

```python
# 위에는 판다스의 어레이
# 아래는 np의 어레이
sequences = np.array(sequences)
X = sequences[: , :-1]
y = sequences[: ,-1]
```

```python
print(X[:3])
```

```python
print(y[:3])
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y, num_classes = vocab_size) # y값을 원핫 인코딩
```

#### 2) 모델 설계하기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM

embedding_dim = 10 # 각각의 단어들의 의미를 담고 있기 때문에 의미를 차원에 담아 둠 (10으로 둔 이유는 시간등...의 문제로 인해)
# embedding_dim이 클 수록 더 세밀한 표현 가능
hidden_units = 128

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(LSTM(hidden_units))
model.add(Dense(vocab_size, activation='softmax')) # softmax: 확률을 구하고 다 더하면 1

# 모델이 학습할 준비
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
# Binary Cross-Entropy (이진 교차 엔트로피)
# Categorical Cross-Entropy (다중 클래스 교차 엔트로피)
model.fit(X, y, epochs = 200, verbose=2)
```

```python
def sentence_generation(model, tokenizer, current_word, n):
    init_word = current_word
    sentence = ''

    for i in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre') # 한단어 썼기 때문에 -1

        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis = 1)
        # 단어집합사전에서의 인덱스값으로 result가 저장됨
        # -> 예측한 단어로 가장 확률이 높은 단어의 인덱스 값

        #tokenizer.word_index -> {단어 : 인덱스, 단어 : 인덱스}
        for word, index in tokenizer.word_index.items(): # .items() : 모든 키와 값이 튜플형태로 나옴
            if index == result:
                break

        current_word = current_word + ' ' + word
        sentence = sentence + ' ' + word
    sentence = init_word + sentence
    return sentence
```

```python
print(sentence_generation(model, tokenizer, 'i', 10))
```

```python
print(sentence_generation(model, tokenizer, 'i', 20))
```

```python

import kagglehub
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

print("Path:", path)
```

```python
del data['Unnamed: 2']
del data['Unnamed: 3']
del data['Unnamed: 4']
```

```python
# ham 과 spam을 0과 1로 변환
data['v1'] = data['v1'].replace(['ham','spam'], [0,1])
data[:5]
```

```python
# 결측치 확인
data.isnull().values.any() # any() : 한번에 True, False로 있는지 없는지 확인
```

```python
data['v2'].nunique()
```

```python
data.drop_duplicates(subset=['v2'], inplace=True)
print(f'총 샘플의 개수 : {len(data)}')
```

```python
data.groupby('v1').size().reset_index(name='count')
```

```python
X_data = data['v2']
y_data = data['v1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=0, stratify=y_data) # stratify=y_data : 0과 1의 수를 균일하게 만들어줌
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)

X_train_encoded = tokenizer.texts_to_sequences(X_train)
print(X_train_encoded[:5])
```

```python
word_to_index = tokenizer.word_index
word_to_index
```

```python
vocab_size = len(word_to_index) + 1
print(f'단어 집합의 크기 : {vocab_size}')
```

```python
max_lengths = 0
for sample in X_train_encoded:
    lengths = len(sample)
    if lengths > max_lengths:
        max_lengths = lengths
print(f'메일의 최대 길이 : {max_lengths}')
```

```python
lengths = [] # 문장의 길이를 담을 리스트
for sample in X_data:
    lengths.append(len(sample))

print(lengths)
```

```python
plt.hist(lengths, bins = 50)
plt.xlabel('lengths')
plt.ylabel('count')
plt.show()
```

```python
X_train_padding = pad_sequences(X_train_encoded, maxlen = max_lengths)
```

```python
X_train_padding # 한의 행에 189개의 열이 들어 감
```

```python
embedding_dim = 8
hidden_units = 32

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer='rmsprop', loss = 'binary_crossentropy', metrics = ['acc']) # metrics = ['acc'] : 분류이기 때문에
history = model.fit(X_train_padding, y_train, epochs = 5, batch_size=64, validation_split=0.2)
```

```python
X_test_encoded = tokenizer.texts_to_sequences(X_test)
X_test_padding = pad_sequences(X_test_encoded, maxlen = max_lengths)
accuracy = model.evaluate(X_test_padding, y_test)[1]
print(accuracy)
```

```python
epochs = range(1, len(history.history['acc']) +1) # loss 값이 확 튀었다가 좋아지면 과접합을 의심해야 함
plt.plot(epochs, history.history['loss'])
plt.plot(epochs, history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','val'])
plt.show()
```
