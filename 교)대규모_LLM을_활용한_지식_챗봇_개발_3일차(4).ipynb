```python
MY_HF_TOKEN = input('ÌÜ†ÌÅ∞ÏùÑ ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî') # ÌÜ†ÌÅ∞ Í¥ÄÎ¶¨ ÏûòÌïòÍ∏∞
```

```python
import transformers
import torch
```

```python
model_id = 'meta-llama/Llama-3.2-1B-Instruct'
```

```python
pipline = transformers.pipeline(
    'text-generation',
    model = model_id,
    model_kwargs = {'torch_dtype' : torch.bfloat16},
    # GPUÏóêÏÑú ÌÅ∞ Î™®Îç∏ ÏÇ¨Ïö©Ïãú Ï£ºÎ°ú ÏÇ¨Ïö©ÎêòÎäî Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ bfloat16
    device_map = 'auto',
    # GPUÏôÄ CPUÎ•º ÏûêÎèôÏúºÎ°ú Ï†ÅÏ†àÌïòÍ≤å ÏÇ¨Ïö©
    token = MY_HF_TOKEN
)
```

```python
messages = [
    {
        'role' : 'system',
        'content' : 'You are a helpful AI assistant who answers question in a friendly and informal manner.'
    },
    {
        'role' : 'user',
        'content' : 'Tell us about Transformers self-attention.'
    }
]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
chat_temp = tokenizer.apply_chat_template(messages, add_generation_prompt = True)
```

```python
tokenizer.decode(chat_temp)
```

```python
terminators = [
    pipline.tokenizer.eos_token_id,
    pipline.tokenizer.convert_tokens_to_ids('<|eot_id|>') # ÎùºÎßàÎ™®Îç∏ÏóêÏÑú ÏÇ¨Ïö©ÌïòÎäî ÏãúÌÄÄÏä§Ïùò ÎßàÏßÄÎßâ
]
```

```python
outputs = pipline(
    messages,
    max_new_tokens = 512,
    eos_token_id = terminators,
    do_sample = True,
    temperature = 0.6,
    # Í∞íÏù¥ ÎÜíÏùÑ ÏàòÎ°ù Ï∞ΩÏùòÏ†Å
    top_p = 0.9
)
```

```python
print(outputs[0]['generated_text'][-1]['content'])
```
